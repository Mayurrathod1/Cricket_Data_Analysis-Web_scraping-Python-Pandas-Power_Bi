# Batting Summary

```python
# Load batting summary JSON data
with open(r'C:\Projects\Cricket t20 team - web scrabing - python - powerbi project\t20_json_files\t20_json_files\t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
```

```python
# Convert batting summary into a DataFrame
df_batting = pd.DataFrame(all_records)

# Display first 5 rows of batting data
df_batting.head()
```

```python
df_batting.tail()
```

```python
# Add a new column 'out/not_out' to categorize players as 'out' or 'not_out'
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
```

```python
# Drop the 'dismissal' column as it's no longer needed
df_batting.drop(columns=["dismissal"], inplace=True)

# Display the updated DataFrame without the 'dismissal' column
df_batting.head()
```

```python
# Cleanup special characters in the batsman names (e.g., replacing 'â€' and '\xa0' with an empty string)
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
match_ids_dict = {}
for index, row in df.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id'] 
```

```python
# Map match names to match IDs using the previously created dictionary
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
```

```python
# Display the DataFrame with the new 'match_id' column
df_batting.head()
```

```python
# Save the cleaned batting summary data to a CSV file
df_batting.to_csv(r'C:\Projects\Cricket t20 team - web scrabing - python - powerbi project\t20_csv_all_files\batting_summary.csv', index=False)
```
